```python
import pingouin as pg
import pandas as pd

# Data Case 1, 2 and 3.
data = pd.read_excel('/home/david/Descargas/expert_results.xlsx', sheet_name='expert_recommendations')

# Select Columns by each nutrient.
columns_N = ['Experto 1 N', 'Experto 2 N', 'Experto 3 N', 'Experto 4 N', 'Experto 5 N', 'Experto 6 N']
columns_P = ['Experto 1 P', 'Experto 2 P', 'Experto 3 P', 'Experto 4 P', 'Experto 5 P', 'Experto 6 P']
columns_K = ['Experto 1 K', 'Experto 2 K', 'Experto 3 K', 'Experto 4 K', 'Experto 5 K', 'Experto 6 K']

# Adjust dataframe
if 'Caso' not in data.columns:
    data['Caso'] = range(1, len(data) + 1)
data = data.dropna(subset=columns_N + columns_P + columns_K)
print('data')
print(data)

# Function to calculate ICCa for a specific column (N, P or K)
def calcular_icca_con_caso(data, columns, tipo):
    data_tipo = data[columns]
    melted_data = data_tipo.melt(var_name='Expert', value_name=tipo)
    melted_data['Caso'] = data['Caso'].repeat(len(columns)).reset_index(drop=True)
    melted_data['Caso'] = melted_data.groupby('Expert').cumcount() + 1 
    print('melted_data para icca_N: ')
    print(melted_data)
    icca = pg.intraclass_corr(data=melted_data, targets='Caso', raters='Expert', ratings=tipo, nan_policy='omit')
    return icca

# Calculate ICCa for N, P and K globally using 'Case' as the target
icca_N = calcular_icca_con_caso(data, columns_N, 'N')
icca_P = calcular_icca_con_caso(data, columns_P, 'P')
icca_K = calcular_icca_con_caso(data, columns_K, 'K')

# Show results
print("ICCa global para N:")
print(icca_N)
print("ICCa global para P:")
print(icca_P)
print("ICCa global para K:")
print(icca_K)
```
